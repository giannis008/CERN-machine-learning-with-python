```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Capture data from "Trace".

```python
from tslearn.datasets import CachedDatasets
X_train, y_train, X_test, y_test = CachedDatasets().load_dataset("Trace")
```

```python
dfs = []
for i in range(len(X_train)):
    dfs.append( pd.DataFrame(X_train[i]).T )
    
```

```python
features = pd.concat(dfs).reset_index(drop=True)
features.head()
```

```python
target = pd.DataFrame(y_train,columns=['label'])
target.head()
```

```python
output = pd.concat([target, features], axis=1)
output.head()
```

```python
output.to_csv('trace.csv',index=False)
```
